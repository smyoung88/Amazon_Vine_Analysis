```python
# Import dependencies
import pandas as pd
import os
import csv
```

```python
# Read in vine_table.csv as a DataFrame
vine_df = pd.read_csv("vine_table.csv")
```

## 1. Filter data where total_votes >= 20 

```python
# Filter data where total_votes >= 20 
vine_votes_df = vine_df[vine_df['total_votes']>=20]
```

```python
vine_votes_df
```

## 2. Filter previous df for rows where helpful_votes/total_votes is >= 50%

```python
# Filter previous df for rows where helpful_votes/total_votes is >= 50%
vine_helpful_df = vine_votes_df[(vine_votes_df['helpful_votes']/vine_votes_df['total_votes'])>=.50]
```

```python
vine_helpful_df
```

## 3. Filter previous df where review was written as part of the Vine program (paid), vine =='Y'

```python
# Filter previous df where review was written as part of the Vine program (paid), vine =='Y'
vine_paid_df = vine_helpful_df[vine_helpful_df['vine']=='Y']
```

```python
vine_paid_df
```

## 4. Filter previous df where review was written not as part of the Vine program (unpaid), vine =='N'

```python
# Filter previous df where review was written not as part of the Vine program (unpaid), vine =='N'
vine_unpaid_df = vine_helpful_df[vine_helpful_df['vine']=='N']
```

```python
vine_unpaid_df
```

### Percentage 5-star reviews for paid

```python
# Determine percentage of 5-star reviews for paid for reviews
# Determine total paid for reviews
total_paid = len(vine_paid_df)
total_paid
```

```python
# Determine total 5 star paid for reviews
total_paid_fivestar = len(vine_paid_df[vine_paid_df['star_rating']==5])
total_paid_fivestar
```
